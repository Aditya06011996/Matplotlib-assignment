 Q 1 What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans - Matplotlib is a popular Python library for creating static, animated, and interactive visualizations in data analysis and scientific computing. It provides a comprehensive set of tools for creating a wide range of plots and charts, making it an essential tool in the data science and scientific communities.

Matplotlib is used for various purposes, including:

(1) Data Visualization: Matplotlib is primarily used for creating data visualizations. It allows users to represent data in various forms, making it easier to understand and interpret complex datasets.

(2) Scientific Research: Scientists and researchers use Matplotlib to visualize experimental data, simulation results, and scientific findings. It helps in presenting data in a graphical format for better communication.

(3) Statistical Analysis: Data analysts and statisticians use Matplotlib to visualize statistical distributions, trends, and relationships within datasets. This aids in making informed decisions based on data.

(4) Educational Purposes: Matplotlib is widely used in teaching and learning data visualization and Python programming. Its simplicity and versatility make it a valuable tool for educational purposes.

(5) Report Generation: Matplotlib can be used to generate charts and plots for reports, presentations, and publications, enhancing the visual appeal of documents.

The Pyplot module of Matplotlib is a collection of functions that provides a simplified interface for creating various types of plots. Here are five common types of plots that can be created using the Pyplot module of Matplotlib:

(1) Line Plot: Line plots are used to visualize the relationship between two continuous variables by connecting data points with lines. They are suitable for displaying trends and patterns over a continuous range.

(2) Scatter Plot: Scatter plots are used to display individual data points as markers on a graph. They are helpful for visualizing the distribution and relationships between two variables.

(3) Bar Plot: Bar plots (or bar charts) are used to represent categorical data with rectangular bars. They are often used to compare quantities or categories.

(4) Histogram: Histograms are used to display the distribution of a single variable by dividing it into bins and counting the frequency of data points within each bin. They are useful for understanding data distributions.

(5) Pie Chart: Pie charts are circular graphs used to represent parts of a whole. They are suitable for showing the composition or proportion of different categories within a dataset.

These are just a few examples of the types of plots that can be created using Matplotlib's Pyplot module. Matplotlib offers a wide range of customization options, allowing users to tailor their plots to meet specific needs.






Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

Ans - A scatter plot is a type of data visualization that displays individual data points as markers on a graph. It is used to visualize the relationship between two variables, allowing you to see how changes in one variable affect the other. Scatter plots are particularly useful for identifying patterns, trends, clusters, or outliers in data.

Here's how you can generate a scatter plot using the provided code and add a title, xlabel, and ylabel to it using Matplotlib:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data for x and y
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Create a scatter plot
plt.scatter(x, y, c='blue', label='Data Points')

# Add title and labels
plt.title('Scatter Plot of x vs. y')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Display the legend (if you want)
plt.legend()

# Show the plot
plt.show()
```

In this code:

*We import NumPy and Matplotlib for data generation and plotting, respectively.
*We generate random data for x and y using NumPy's random.normal function.
*We create a scatter plot using plt.scatter() and specify the color of the markers as blue. You can customize the marker style and color as needed.
*We add a title to the plot using plt.title().
*We label the x-axis and y-axis using plt.xlabel() and plt.ylabel().
*You can also add a legend using plt.legend() if multiple data series are present, though in this case, we have only one.
*Finally, we display the plot using plt.show().

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans - The subplot() function in Matplotlib is used to create multiple subplots or small individual plots within a single figure. It allows you to arrange and display multiple plots in a grid-like fashion, which can be helpful for comparing different datasets, visualizing multiple aspects of data, or presenting related information together.

The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot. Here's how it's used:

```python
plt.subplot(rows, columns, index)
```

Now, let's draw four line plots using the subplot() function with the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data for line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Data for line 1
y1 = np.array([0, 100, 200, 300, 400, 500])

# Data for line 2
y2 = np.array([50, 20, 40, 20, 60, 70])

# Data for line 3
y3 = np.array([10, 20, 30, 40, 50, 60])

# Data for line 4
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a 2x2 grid of subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1)
plt.title('Line 1')

plt.subplot(2, 2, 2)
plt.plot(x, y2)
plt.title('Line 2')

plt.subplot(2, 2, 3)
plt.plot(x, y3)
plt.title('Line 3')

plt.subplot(2, 2, 4)
plt.plot(x, y4)
plt.title('Line 4')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

In this code:

*We import NumPy and Matplotlib for data generation and plotting, respectively.
*We create four sets of data (y1, y2, y3, and y4) and specify x for all of them.
*We use the subplot() function to create a 2x2 grid of subplots. The first subplot is in the top-left position, the second in the top-right, the third in the bottom-left, and the fourth in the bottom-right.
*We plot the respective data on each subplot and add titles to them.
*plt.tight_layout() is used to adjust the spacing between subplots for better readability.
Finally, we display the plots using plt.show().

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

Ans - 
A bar plot, also known as a bar chart or bar graph, is a type of data visualization that uses rectangular bars or columns to represent and compare different categories or groups of data. The length or height of each bar is proportional to the value it represents, making it a useful tool for displaying discrete data or categorical variables.

Bar plots are used for several purposes, including:

(1) Comparing Categories: Bar plots are effective for comparing the values of different categories or groups at a glance. They make it easy to identify which category has the highest or lowest values.

(2) Showing Distribution: Bar plots can display the distribution of data within categories, especially when used as stacked or grouped bars.

(3) Visualizing Frequency: They can represent the frequency or count of items in different categories, making them suitable for displaying histograms or frequency distributions.

(4) Tracking Changes Over Time: When used as a time series bar plot, they can show changes in data over different time intervals or periods.

Now, let's create a bar plot and a horizontal bar plot using the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Create a vertical bar plot
plt.figure(figsize=(8, 4))  # Optional: Adjust the figure size
plt.bar(company, profit, color='skyblue')
plt.xlabel('Company')
plt.ylabel('Profit (in billions)')
plt.title('Bar Plot of Company Profits')
plt.show()

# Create a horizontal bar plot
plt.figure(figsize=(8, 4))  # Optional: Adjust the figure size
plt.barh(company, profit, color='lightcoral')
plt.xlabel('Profit (in billions)')
plt.ylabel('Company')
plt.title('Horizontal Bar Plot of Company Profits')
plt.show()
```

In this code:

We import NumPy and Matplotlib for data handling and plotting.
We define the data for the companies and their profits.
First, we create a vertical bar plot using plt.bar(), specifying the company names on the x-axis and profit values on the y-axis. We add labels, a title, and specify the bar color.
Second, we create a horizontal bar plot using plt.barh(), which is similar to the vertical bar plot but with the axes flipped.
Running this code will generate both a vertical and a horizontal bar plot for the given company profit data.

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

Ans - A box plot, also known as a box-and-whisker plot, is a graphical representation of the distribution of a dataset. It is particularly useful for visualizing the central tendency, spread, and skewness of the data, as well as identifying outliers. Box plots are used for the following purposes:

(1) Summarizing Data: Box plots provide a compact and informative summary of the distribution of a dataset, including the median (center), quartiles (spread), and potential outliers.

(2) Comparing Distributions: They are useful for comparing the distributions of multiple datasets side by side, making it easy to identify differences in central tendency and spread.

(3) Outlier Detection: Box plots highlight outliers, which are data points that fall significantly above or below the central portion of the data distribution.

(4) Identifying Skewness: Box plots can reveal whether the data is symmetric or skewed to one side. Skewed data will have an asymmetrical box plot.

Now, let's create a box plot for the given data:

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Create a box plot
plt.figure(figsize=(8, 6))  # Optional: Adjust the figure size
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], notch=True, patch_artist=True,
            boxprops=dict(facecolor='lightblue', color='blue'),
            medianprops=dict(color='black'))
plt.xlabel('Boxes')
plt.ylabel('Values')
plt.title('Box Plot of Box 1 and Box 2')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```
